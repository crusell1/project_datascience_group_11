Chosen websites: https://www.ibm.com/think/topics/machine-learning, https://mitsloan.mit.edu/ideas-made-to-matter/machine-learning-explained 

```python
# Importeringar
from bs4 import BeautifulSoup
import requests
import sys
```

 Hemsida 1

```python
# Visar vem som skrapar
USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"

# Anger den som skrapar i https-förfrågan
headers = {"user-agent": USER_AGENT}
# Skickar förfrågan till webplatsen
r = requests.get("https://www.ibm.com/think/topics/machine-learning", headers=headers)

# Visar statuskoden från förfrågan
print("Status code: ", r.status_code)
```

```python
# Om statuskoden är 200, dvs okej
if r.status_code == 200:
    # Skriver ut att allt är ok
    print("Scrapping is possible")
    # Hämtar och parsar htmlkoden
    soup = BeautifulSoup(r.content, "html.parser")
else:
    # Skriver ut att det inte är möjligt att skrapa
    print("Scrapping is not possible")
    # Avslutar
    sys.exit(0)
```

```python
# Benämde filen som texten ska lagras i
out_filename = "ibm.txt"

# Hittar titelns tagg, dvs h1
title_tag = soup.find("h1")
# Hämtar all text i titelns tagg och skalar den
# Om ingen titel tagg finns så returneras "Ingen Titel"
title = title_tag.get_text(strip=True) if title_tag else "No Title"

# Hittar div:en med klassen "body-article-8" som består av all text på sidan
content_div = soup.find("div", {"class": "body-article-8"})

# Om div:en med all text inte finns, skickas följande felmeddelande
if not content_div:
    print("Content not found. Website structure may have changed.")
    sys.exit(0)

# Hittar samtliga p- och li-taggar i div:en med all text
paragraphs = content_div.find_all(["p", "li"])

# Initierar variabeln som ska hålla all text
article_text = ""

# Loop för varje text-tagg
for p in paragraphs:
    # Hämtar texten från text-taggen och skalar den
    text = p.get_text(strip=True)
    # Om det finns text, lägg till den och en ny rad i variabeln som håller texten
    if text:
        article_text += text + "\n"

# Öppnar textfilen, väljer skrivläget och specificerar Unicode
with open(out_filename, "w", encoding="utf-8") as f:
    # Skriver in titeln och två nya rader
    f.write(title + "\n\n")
    # Skriver in all text från artikeln
    f.write(article_text)

# Skriver ut meddelande om att det sparats i textfilen
print(f"Content saved to {out_filename}")

# Stänger filen
f.close()
```

```python
# Visar vem som skrapar
USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"

# Anger den som skrapar i https-förfrågan
headers = {"user-agent": USER_AGENT}
# Skickar förfrågan till webplatsen
r = requests.get("https://www.kicks.se/parfym?search-query=parfym", headers=headers)

# Visar statuskoden från förfrågan
print("Status code: ", r.status_code)
```

```python
# Om statuskoden är 200, dvs okej
if r.status_code == 200:
    # Skriver ut att allt är ok
    print("Scrapping is possible")
    # Hämtar och parsar htmlkoden
    soup = BeautifulSoup(r.content, "html.parser")
else:
    # Skriver ut att det inte är möjligt att skrapa
    print("Scrapping is not possible")
    # Avslutar
    sys.exit(0)
```

```python
# Hämtar hela produktlistan
containers = soup.find("div", {"class": "ProductList__Grid-sc-1kxgi4x-0 gdBrrB"})

# Hittar respektive produkt i litan med produkter
container = containers.find_all("div", {"class": "ProductItem__Wrapper-sc-1wmia3i-0 bQprsQ"})

# Benämde csv-filen
out_filename = "kicks.csv"
# Anger kolumnnamen i csv-filen
headers = "Varumärke, Namn\n"
```

```python
# Öppnar filen, anger möjlighet att skriva, läsa och uppdatera
f = open(out_filename, "w+")
# Skriver in kolumnnamnen
f.write(headers)

# Loop för varje produkt i produktlistan
for con in container:

    # Hämtar allt innehåll i varje produkt
    anchors = con.find_all('div', {"class": "CardContent__StyledCardContent-sc-15yb5cd-0 hjpqjg styles__ProductContent-sc-1hwmg7-1 dAXpFB"})
    # Om det finns innehåll
    if anchors:

        # Hämtar produktens varumärke
        brand = con.find("span", {"class": "Label__StyledLabel-sc-tlgbnf-0 bOpeyU styles__Brand-sc-1ijnrjd-0 jTklaf"})
        # Hämtar produktens namn
        name = con.find('span', {"class" : "Body__StyledBody-sc-1bcfcvm-0 eHlofd styles__Product-sc-1ijnrjd-1 dIsTUL"})
            
        # Sparar produktens attribut i textform i variabler
        b = brand.text
        c = name.text

        # Skriver ut attributen i textform
        print(b, c)

        # Lägger till attributen i textform i csv-filen
        # GEMINI START, början av första produkten "försvann" i csv-filen
        f.write(f"{b}, {c}\n")
        # GEMINI SLUT

# Stänger filen
f.close()
```
