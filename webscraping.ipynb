```python
# Importeringar
from bs4 import BeautifulSoup
import requests
import sys
import time
```

```python
# Visar vem som skrapar
USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"

# Anger den som skrapar i https-förfrågan
headers = {"user-agent": USER_AGENT}
# Skickar förfrågan till webplatsen
r = requests.get("https://www.systembolaget.se/sitemap-produkter-ol.xml", headers=headers)

# Visar statuskoden från förfrågan
print("Status code: ", r.status_code)
```

```python
# Om statuskoden är 200, dvs okej
if r.status_code == 200:
    # Skriver ut att allt är ok
    print("Scrapping is possible")
    # Hämtar och parsar xmlkoden
    soup = BeautifulSoup(r.content, "xml")
    print(soup)
else:
    # Skriver ut att det inte är möjligt att skrapa
    print("Scrapping is not possible")
    # Avslutar
    sys.exit(0)
```

```python
# Hittar url:ernas loc-tagg
loc_tags = soup.find_all("loc")

# Initierar en array för att hålla alla produkters url
product_urls = []

# GEMINI: start
# För varje tag i loc_tags
for tag in loc_tags:
  # Hämtar och skalar url:en från respektive tag
  url = tag.get_text(strip=True)
  # Lägger till url:en i arrayen med länkar
  product_urls.append(url)
# GEMINI: slut

# Bekräftar att vi fått korrekt antal länkar (finns 4837 varor i ölkategorin)
print(len(product_urls))
```

```python
for url in product_urls:
  productRequest = requests.get(url=url, headers=headers)

  # Om statuskoden är 200, dvs okej
  if productRequest.status_code == 200:
      # Hämtar och parsar htmlkoden
      soupProducts = BeautifulSoup(productRequest.content, "html.parser")
  else:
      # Skriver ut att det inte är möjligt att skrapa
      print("Scrapping is not possible")
      # Avslutar
      sys.exit(0)

  # Väntar 2 sekunder för att inte få detta att se ut som en attack
  time.sleep(2)
```
